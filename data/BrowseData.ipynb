```python
import json
import requests
import os.path as osp
import pandas as pd
from termcolor import colored
from botchat.utils import load, dump

def load_tsv(f):
    return pd.read_csv(f, sep='\t')

def download_file(url, filename=None):
    if filename is None:
        filename = url.split('/')[-1]
    response = requests.get(url)
    open(filename, 'wb').write(response.content)
    
def print_dlg(dlg):
    dlg = json.loads(dlg)
    chars = 'AB' * 999
    lines = []
    for c, u in zip(chars, dlg):
        s = f'{c}: {u}' 
        lines.append(colored(s, 'red' if c == 'A' else 'blue'))
    print('\n'.join(lines))
```

```python
conv_url = 'https://opencompass.openxlab.space/utils/BotChat/MuTualTest-convs.tsv'
start_url = 'https://opencompass.openxlab.space/utils/BotChat/MuTualTest-starts.tsv'
if not osp.exists(conv_url.split('/')[-1]):
    download_file(conv_url)
if not osp.exists(start_url.split('/')[-1]):
    download_file(start_url)
```

```python
from tiktoken import encoding_for_model
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
enc = encoding_for_model('gpt-4-0613')

def lt_stats(data):
    models = list(data.keys())
    models.remove('index')
    models.remove('nround')
    nq = len(data)
    nm = len(models)
    ndarr = np.zeros((nm, nq, 14), dtype=int)
    for j in range(nq):
        item = data.iloc[j]
        for i, m in enumerate(models):
            dlg = json.loads(item[m])
            for k, words in enumerate(dlg[2:]):
                ndarr[i, j, k] = len(enc.encode(words))
    return models, ndarr

def build_df(models, ndarr):
    res = dict(model=[], length=[])
    for i, m in enumerate(models):
        for j in range(547):
            for k in range(14):
                res['model'].append(m)
                res['length'].append(np.log10(max(ndarr[i][j][k], 1)))
    return pd.DataFrame(res)

data = load('MuTualTest-convs.tsv')

models, ndarr = lt_stats(data)
df = build_df(models, ndarr)

fig, ax = plt.subplots(figsize=(16,6))

sns.violinplot(x=df['model'], y=df['length'])

params = {'mathtext.default': 'regular'}          
plt.rcParams.update(params)

ax.set_xlabel('')

labels = ax.get_xticklabels()
model_names = [x.get_text() for x in labels]
ax.set_xticklabels(model_names, rotation=30)

ax.set_ylabel('Utterance Token Length (CL100K)')
labels = ax.get_yticklabels()
yticks = [x.get_text() for x in labels]
yticks = [x[0] if '.0' in x else x for x in yticks]

ax.set_yticklabels(['$10^{' + x + '}$' for x in yticks])

fig.tight_layout()
fig
```
